```python
# read csv file using pandas
import pandas as pd

# read csv file using pandas
data = pd.read_csv("~/winhome/Downloads/197 Industry Groups.csv")
data.head()

# drop columns "Order" and "Symbol" from data
data = data.drop(columns=["Order", "Symbol", "Number of Stocks"])

# rename column "Group" to "Group Name"
data = data.rename(columns={"Ind Group Rank": "Rnk Actual", "Ind Grp Rnk Last Week": "Rnk Last Week", "Ind Grp Rnk 3 Mo Ago": "Rnk 3 Mo Ago", "Ind Grp Rnk 6 Mo Ago": "Rnk 6 Mo Ago"})

# create a new column in data called "Diff" calculated by "Rnk Actual" - "Rnk 3 Mo Ago"
data["Diff3"] = data["Rnk Actual"] - data["Rnk 3 Mo Ago"]
data["Diff6"] = data["Rnk Actual"] - data["Rnk 6 Mo Ago"]

data.head(30)
```
